# IMPORTING NECESSARY LIBRARIES

```python
import pandas as pd 
import numpy as np
from sklearn import linear_model
import seaborn as sns
```

# ANALYSING DATA

```python
car=pd.read_csv("https://raw.githubusercontent.com/amankharwal/Website-data/master/CarPrice.csv")
car.head(100)
```

```python
car.info()
```

```python
car.describe()
```

```python
carr=car.drop(['CarName','symboling','fueltype','aspiration','doornumber','fuelsystem','drivewheel','enginelocation'],axis=1)
target=car.price
print(carr)
print()
print("Price column:")
print()
print(target)
```

```python
from sklearn.preprocessing import LabelEncoder
Numerics=LabelEncoder()

carr['carbody_n'] = Numerics.fit_transform(carr['carbody'])
carr['enginetype']= Numerics.fit_transform(carr['enginetype'])
carr['cylindernumber']= Numerics.fit_transform(carr['cylindernumber'])

carr
```

```python
carr.columns
```

# DATA SPLIT

```python
X=carr[['enginetype', 'horsepower','enginesize', 'highwaympg','horsepower','highwaympg','peakrpm','citympg','carheight','carwidth'
    ,'curbweight','car_ID','wheelbase','carbody_n','compressionratio','carlength','boreratio','stroke','cylindernumber']]
y=carr[['price']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=7002)
```

# MODEL_CREATION AND TRAINING  

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
print("Accuracy is:",model.score(X_test,y_test)*100)
```
