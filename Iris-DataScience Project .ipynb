#  IRIS FLOWER CLASSIFICATION

This iris flower classification is an example of multivariate used for classification task.The machine learning model can be trained using several algorithms and techniques to accurately predict the new sample of the iris flower according to their attribute such as SepalLength,SepalWidth,PetalLength,PetalWidth.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(style="white",color_codes=True)
```

```python
iris=pd.read_csv("Iris.csv")
iris.head()
```

```python
iris["Species"].value_counts()
```

Scatter Plot

```python
iris=iris.drop(columns=['Id'])
iris.head()
```

# PAIRPLOT USING SEABORN

```python
penguins = sns.load_dataset("iris")
sns.pairplot(iris)
```

```python
flower_mapping={'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2}
iris["Species"]=iris["Species"].map(flower_mapping)
```

```python
iris.head(150)
```

# ACCURACY

```python
from sklearn.model_selection import train_test_split
x=iris.drop(columns=['Species'])
y=iris['Species']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

Accuracy is 93%
